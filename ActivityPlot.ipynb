# Analyse de l'activité des paquets et dépôts R

Ce notebook a pour but de fournir quelques étapes (et le code associé) permettant de visualiser l'activité des dépôts et des paquets R. Dans un premier temps, nous chargeons les noms et propriétaires des dépôts R qui correspondent à un package R (autrement dit : qui ont un fichier `DESCRIPTION`). Ensuite, nous chargeons la liste des événéments du flux d'activités de chacun de ces dépôts, et nous comparons par groupe : les dépôts qui ne sont pas des packages, les dépôts qui sont des packages, les dépôts qui sont des packages SAUF ceux de cran et rpkg (deux *owner* particuliers).

## Chargement des données depuis les fichiers "description" des packages R

```python
import pandas
import numpy

%matplotlib inline

# Load from Maelick's data
df_description = pandas.read_csv('/data/github-description.csv')
```

```python
df_description.head()
```

```python
df_packages = df_description[['owner', 'repository']].drop_duplicates()
df_packages.describe()
```

```python
# One could simply add labels for 'in_cran', 'in_rpkg', and 'in_other' for df_packages...
df_packages_cran = df_packages.query('owner == "cran"').copy()
df_packages_rpkg = df_packages.query('owner == "rpkg"').copy()
df_packages_others = df_packages.query('owner != "cran" and owner != "rpkg"').copy()
```

 - rpkg possède 4409 dépôts (dont 4344 hébergent un package commun avec un dépôt cran)

## Chargement du flux d'activités

Nous allons extraire les informations du flux d'activités. En particulier, nous garderons :
- Le nom et le propriétaire du dépôt, 
- Le type de l'événement, 
- La date de l'événement

```python
import pymongo
import dateutil.parser
from itertools import imap

conn = pymongo.MongoClient()

def _from_json_to_flat(e):
    return {'type': e['type'], 
                   'created_at': dateutil.parser.parse(e['created_at']),
                   'owner': e['repository']['owner'],
                   'repository': e['repository']['name']}

events = imap(_from_json_to_flat, 
              conn.r.events.find({'type': 'PushEvent'},
                                 fields=['type', 'created_at', 'repository.name', 'repository.owner']))
```

```python
# This should take some time (expect 683425 items)
df_events = pandas.DataFrame(list(events))
```

```python
# Add flag to every packages in cran, rpkg, etc. 
df_packages_cran['cran'] = 1
df_packages_rpkg['rpkg'] = 1
df_packages_others['other'] = 1

# Merge them all!
df_events_all = pandas.merge(df_events, df_packages_cran, how='left', on=('owner', 'repository'))
df_events_all = pandas.merge(df_events_all, df_packages_rpkg, how='left', on=('owner', 'repository'))
df_events_all = pandas.merge(df_events_all, df_packages_others, how='left', on=('owner', 'repository'))

# Date indexed
df_events_all.set_index('created_at', inplace=True)

df_events_all['all'] = 1
```

## Graphiquement...

Le graphique suivant reprend une moyenne glissante sur 30 jours du nombre d'événements de type "PushEvent". Chaque courbe correspond à une partie (ou le tout) des dépôts :
- la courbe bleue correspond aux dépôts contenant un package R lié au compte cran, 
- la courbe verte est invisible, mais elle correspond à un tout petit trait aux alentours d'avril 2014 (un peu avant le début de la courbe bleue) et correspond aux dépôts contenant un package R lié à rpkg,
- la courbe rouge correspond aux dépôts contenant un package R, qui n'appartient ni à rpkg, ni à cran.
- la courbe cyan correspond à l'ensemble de tous les dépôts liés à R.

```python
df_temp = df_events_all.resample('1D', label='left', loffset='1d', how='sum')
plot = pandas.stats.moments.rolling_mean(df_temp, window=30)[['cran', 'rpkg', 'all', 'other']].plot()
plot.set_title(u'Événéments par catégorie de dépôts')
plot.set_xlabel('date')
plot.set_ylabel(u'# événements')
plot.get_figure().set_size_inches(15,6)
```

Ou encore, l'*accélération* : 

```python
plot = pandas.stats.moments.expanding_mean(df_temp)[['cran', 'rpkg', 'all', 'other']].plot()
plot.set_title=(u'Accélération du nombre d\'événements par catégorie de dépôts')
plot.set_ybound(upper=1000)
plot.set_xlabel('date')
plot.set_ylabel(u'moyenne cumulée')
plot.get_figure().set_size_inches(15,6)
```

## Activité des dépôts contenant un package sur CRAN et en dehors

Nous allons utiliser les informations collectées depuis CRAN afin d'identifier quels sont les packages R disponibles sur Github qui sont également présents sur CRAN. Cela nous permettra de comparer l'activité des dépôts qui sont liés à un package "officiel" de ceux qui ne le sont pas. 

Pour procéder à cette identification, nous comparons le champ *Package* des fichiers `DESCRIPTION` des packages R de Github avec ceux de CRAN.

```python
# Load again from Maelick's data
df_description = pandas.read_csv('/data/github-description.csv')
df_cran = pandas.read_csv('/data/cran-description.csv')
```

```python
df_description = (df_description[df_description['key'] == 'Package']
                  .drop('key', axis=1)
                  .rename(columns={'value': 'package'}))

df_cran = (df_cran
           .drop_duplicates('package')
           .drop(['version', 'key', 'value'], axis=1))

df_cran['on_CRAN'] = 1
```

```python
# Merge and fill NaN with "0" for 'on_CRAN'
df_packages = df_description.merge(df_cran, how='left', on='package').fillna(0)
df_packages_events = df_packages.merge(df_events, how='inner', on=('owner', 'repository'))

# Prepare columns for easy plotting
df_packages_events.set_index('created_at', inplace=True)
df_packages_events['all'] = 1
df_packages_events['not_on_CRAN'] = 1 - df_packages_events['on_CRAN']
```

```python
df_temp = df_packages_events.resample('1D', label='left', loffset='1d', how='sum')
plot = pandas.stats.moments.rolling_mean(df_temp, window=30)[['on_CRAN', 'all', 'not_on_CRAN']].plot()
plot.set_title(u'Événéments par type de packages')
plot.set_xlabel('date')
plot.set_ylabel(u'# événements')
plot.get_figure().set_size_inches(15,6)
```

Le petit "pic" en avril peut sans doute d'expliquer par l'import des packages depuis CRAN vers le compte cran sur Github (mirroring des packages). On peut filtrer cela.

```python
df_temp = df_packages_events[df_packages_events['owner'] != 'cran'].resample('1D', label='left', loffset='1d', how='sum')
plot = pandas.stats.moments.rolling_mean(df_temp, window=30)[['on_CRAN', 'all', 'not_on_CRAN']].plot()
plot.set_title(u'Événéments par type de packages')
plot.set_xlabel('date')
plot.set_ylabel(u'# événements')
plot.get_figure().set_size_inches(15,6)
```
