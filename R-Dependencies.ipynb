# Analyse des dépendances des paquets R sur CRAN et GitHub

Nous allons analyser les dépendances des différents paquets R présents sur CRAN et sur GitHub, afin de voir s'il est possible d'identifier une différence de comportement entre ceux qui sont uniquement développés/publiés sur GitHub et ceux qui sont uniquement développés/publiés sur CRAN. 

Ce notebook nécessite d'avoir accès aux fichiers `DESCRIPTION` des paquets R, qu'ils soient de CRAN ou de GitHub.

```python
import pandas

# Load data from Maelick's files
github = pandas.read_csv('github-description.csv')
cran = pandas.read_csv('cran-description.csv')

# Drop packages belonging to "cran" user on GitHub
github = github[github['owner'] != 'cran']
```

```python
_ = github.query('key == "Package"').rename(columns={'value': 'package'})
_ = _.merge(github, how='left', on=('owner', 'repository'))[['package', 'key_y', 'value']]
github = _.rename(columns={'key_y': 'key'}).drop_duplicates(('package', 'key'), take_last=True)
```

```python
github_pkg = github.pivot(index='package', columns='key', values='value')[['Depends', 'Imports']]
cran_pkg = cran.drop_duplicates(('package', 'key'), take_last=True).pivot(index='package', columns='key', values='value')[['Depends', 'Imports']]
```

Now, `github_dep` and `cran_dep` have the same structure: package name, list of dependencies, list of imports. 
Here is what does `Imports` or `Depends` look like:

```python
cran_pkg.tail()
```

```python
# Add a flag to both data frame
cran_pkg['cran'] = 1
github_pkg['github'] = 1

# Combine the two data frames, using CRAN data at first
pkg = cran_pkg.combine_first(github_pkg)
```

We need to parse `Depends` and `Imports` to extract the related packages. We only consider the name of the package (not its version), i.e. a contiguous sequence of characters, so it is easily expressed as follows:

```python
def parse_packages_list(pkg_list):
    return map(str.strip, map(lambda p: p.split('(')[0], pkg_list.split(',')))
```

```python
_ = cran_pkg.loc['zooimage']['Depends']
print 'Example:', _
print 'Result:', parse_packages_list(_)
```

```python
# Replace every NaN in Depends and Imports by an empty string (i.e. an empty list of packages)
pkg = pkg.fillna(value={'Depends': '', 'Imports': ''})

# Count 'Depends' and 'Imports' items
pkg['dependencies'] = pkg['Depends'].map(lambda p: len(parse_packages_list(p))) + pkg['Imports'].map(lambda p: len(parse_packages_list(p)))
```

Some reshaping...

```python
pkg['cran_dep'] = pkg['dependencies'] * pkg['cran']
pkg['github_dep'] = pkg['dependencies'] * pkg['github']
pkg['both_dep'] = pkg['dependencies'] * pkg['cran'] * pkg['github']

pkg['cran_only'] = pkg.apply(lambda x: x['dependencies'] if x['cran'] == 1 and x['github'] != 1 else pandas.np.nan, axis=1)
pkg['github_only'] = pkg.apply(lambda x: x['dependencies'] if x['cran'] != 1 and x['github'] == 1 else pandas.np.nan, axis=1)
```

## Graphiquement...

```python
%matplotlib inline
pandas.options.display.mpl_style = 'default'
```

```python
columns = ['cran_dep', 'github_dep', 'both_dep', 'cran_only', 'github_only']

pkg[columns].plot(kind='box', logy=True, figsize=(15,6), title=u'Dépendances par paquet')
```

```python
pkg_hist = pandas.concat([pandas.value_counts(pkg[c]) for c in columns], axis=1)
pkg_hist = pkg_hist.rename(columns=dict(enumerate(columns)))
pkg_hist.plot(kind='bar', figsize=(15,6), xlim=(-1,10), title=u'Nombre de paquets par nombre de dépendances')
```

```python
pandas.tools.plotting.scatter_matrix(pkg_hist, marker='o', figsize=(15,15), diagonal='kde', grid=True)
```
