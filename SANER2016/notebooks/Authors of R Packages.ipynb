```python
data = pandas.DataFrame.from_csv('../data/github-raw-150601.csv', index_col=None)
data = data.drop(labels='Unnamed: 0', axis=1).dropna(subset=('Package', 'Version'))
data['CommitDate'] = pandas.to_datetime(data['CommitDate'])
```

We remove the packages that are owned by BioConductor-mirror, rpkg and cran, as they are mainly mirrored from (resp.) BioConductor and CRAN (twice).

```python
ndata = data.query('Owner == "Bioconductor-mirror" or Owner == "rpkg" or Owner == "cran"')
print 'This affects', len(ndata), 'rows out of', len(data)
data = data.drop(ndata.index)
```
