# Analyse des dépendances des paquets R

Nous allons analyser les dépendances des différents paquets R que nous avons collectés depuis CRAN, BioConductor, R-Forge, RForge et GitHub afin de voir des éventuelles variations sur le nombre ou la provenance des dépendances.

```python
import pandas

# Load data from Maelick's files
github = pandas.read_csv('../data/github-description.csv')
cran = pandas.read_csv('../data/cran-description.csv')
bioconductor = pandas.read_csv('../data/bioconductor_description.csv')
r_forge = pandas.read_csv('../data/r-forge_description.csv')
rforge = pandas.read_csv('../data/rforge_description.csv')
```

```python
# Drop packages belonging to "cran" user on GitHub
github = github[github['owner'] != 'cran']

# Pivot on key->value
_packages = github[github['key'] == 'Package'].rename(columns={'value': 'Package'})[['owner', 'repository', 'Package']]
_depends = github[github['key'] == 'Depends'].rename(columns={'value': 'Depends'})
_imports = github[github['key'] == 'Imports'].rename(columns={'value': 'Imports'})
_github = _packages.merge(_depends, how='left', on=('owner', 'repository')).merge(_imports, on=('owner', 'repository'))
github = _github.set_index('Package')[['Imports', 'Depends']]

# Indexes
bioconductor = bioconductor.rename(columns={'Unnamed: 0': 'Package'}).set_index('Package')[['Depends', 'Imports']]
rforge = rforge.set_index('Package')[['Depends', 'Imports']]
r_forge = r_forge.set_index('Package')[['Depends', 'Imports']]

# Pivot & index
_cran = cran.drop_duplicates(('package', 'key'), take_last=True)
_depends = _cran[_cran['key'] == 'Depends'].rename(columns={'value': 'Depends'})
_imports = _cran[_cran['key'] == 'Imports'].rename(columns={'value': 'Imports'})
_cran = _cran.merge(_depends, how='left', on='package').merge(_imports, how='left', on='package')
cran = _cran.rename(columns={'package': 'Package'}).drop_duplicates('Package').set_index('Package')[['Imports', 'Depends']]
```

Certains paquets sont livrés "par défaut" avec R. Ces paquets vont être repris dans un DataFrame virtuel `R`, sans dépendance.

```python
_ = ('R MASS Matrix base boot class cluster codetools compiler datasets foreign grDevices ' +
        'graphics grid lattice manipulate methods mgcv nlme nnet parallel rpart rstudio' +
        'spatial splines stats stats4 stringr survival tcltk tools translations utils').split(' ')
R = {}
for pkg in _:
    R[pkg] = { 'Depends': '', 'Imports': '' }
R = pandas.DataFrame.from_dict(R, orient='index')
```

Une petite fonction pour parser les dépendences et les retourner sous forme de liste.

```python
def parse_dependencies(item):
    depends = item['Depends']
    imports = item['Imports']
    f = lambda lst: [dep.split('(')[0].strip() for dep in lst.split(',')]
    return filter(lambda x: len(x) > 0, f(depends) + f(imports))
```

On prépare la future liste des packages. Cette liste contiendra la source du package (sa communauté), ainsi que les différentes dépendances. Les sources sont prises dans un certain ordre, faites attention à cela.

```python
packages = {}

sources = [('R', R),
           ('cran', cran), 
           ('bioconductor', bioconductor),
           ('r-forge', r_forge),
           ('rforge', rforge),
           ('github', github)]
            
for source_name, source in sources:
    for name, serie in source.fillna('').iterrows():
        packages.setdefault(name, {'source': source_name, 'deps': parse_dependencies(serie)})
del packages[pandas.np.nan]

sources.append(('Unknown', None))
```

On prépare le dictionnaire de dépendences. Ce dictionnaire va contenir, pour chaque source, les packages concernés par une dépendance vers chaque source. Le dictionnaire contient aussi les dépendances concernées.  

```python
dependencies = {}

for name, source in sources:
    d = {}
    for _name, _source in sources:
        d[_name] = {'packages': [], 'deps': []}
    dependencies[name] = d
```

```python
for name, value in packages.iteritems():
    pkg = {}
    for depname in value['deps']:
        dep = packages.get(depname, {'source': 'Unknown', 'deps': []})
        dependencies[value['source']][dep['source']]['deps'].append(depname)
        pkg[dep['source']] = True
    for source in pkg.iterkeys():
        dependencies[value['source']][source]['packages'].append(name)
```
